```python
prices = {
    'GS': GS['close'],
    'META': META['close'],
    'EURUSD': EURUSD['close'],
    'TCS': TCS['close'],
    'NIFTY50': NIFTY50['close'],
    'NIFTYBANK': NIFTYBANK['close'],
    'RELIANCE': RELIANCE['close'],
    'TATAMOTORS': TATAMOTORS['close'],
    'AAPL': AAPL['close'],
    'AMZN': AMZN['close'],
    'AUDUSD': AUDUSD['close'],
    'BNBUSDT': BNBUSDT['close'],
    'DOGEUSDT': DOGEUSDT['close'],
    'BTCUSDT': BTCUSDT['close'],
    'SOLUSDT': SOLUSDT['close'],
    'DOW': DOW['close'],
    'GBPUSD': GBPUSD['close'],
    'USDCAD': USDCAD['close'],
    'USDJPY': USDJPY['close'],
    'ETHUSDT': ETHUSDT['close'],
    'MSFT': MSFT['close'],
    'GODREJIND': GODREJIND['close']
}

max_length = max(len(v) for v in prices.values())

for key in prices:
    while len(prices[key]) < max_length:
        prices[key] = pd.concat([prices[key], pd.Series([np.nan])], ignore_index=True)

price_df = pd.DataFrame(prices)
print(price_df)

returns = price_df.pct_change().dropna()
```

```python
returns_dict={}
for df in [GS, META, ETHUSDT, EURUSD, TCS, NIFTY50, NIFTYBANK, RELIANCE, TATAMOTORS, AAPL, AMZN, AUDUSD, BNBUSDT, DOGEUSDT, BTCUSDT, SOLUSDT, DOW, GBPUSD, USDCAD, USDJPY, MSFT, GODREJIND]:
    returns_neo=df['close'].pct_change().dropna()
    returns_dict[f"{df}"]=returns_neo
returns_df=pd.DataFrame(returns_dict)

```

```python
expected_returns = returns_df.mean()
covariance_matrix = returns_df.cov()
print(expected_returns)
```

```python
from scipy.optimize import minimize

def portfolio_performance(weights, expected_returns, covariance_matrix):
    portfolio_return = np.dot(weights, expected_returns) * 252  # Annualized return
    portfolio_risk = np.sqrt(np.dot(weights.T, np.dot(covariance_matrix, weights))) * np.sqrt(252)  # Annualized risk
    return portfolio_return, portfolio_risk

def min_sharpe(weights):
    portfolio_return, portfolio_risk = portfolio_performance(weights, expected_returns, covariance_matrix)
    return -portfolio_return / portfolio_risk

constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
bounds = tuple((0, 1) for _ in range(len(expected_returns)))

initial_weights = [1 / len(expected_returns)] * len(expected_returns)

optimal_weights = minimize(min_sharpe, initial_weights, method='SLSQP', bounds=bounds, constraints=constraints)

optimal_weights = optimal_weights.x
```

```python
optimal_return, optimal_risk = portfolio_performance(optimal_weights, expected_returns, covariance_matrix)


print("Optimal Weights:")
for asset, weight in zip(price_df.columns, optimal_weights):
    print(f"{asset}: {weight:.2%}")

print(f"\nExpected Annual Return: {optimal_return:.2%}")
print(f"Annual Risk (Volatility): {optimal_risk:.2%}")
```
